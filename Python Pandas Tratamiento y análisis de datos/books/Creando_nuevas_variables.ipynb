```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_sin_not_null.csv', sep = ';')
```

```python
datos.head(10)
```

```python
#crando variable con espacios, suma de 3 valores ejemplo
#datos['Valor Bruto'] = () + () + ()
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
```

```python
#valor por metro cuadrado
datos['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head(10)
```

```python
#valor por metro cuadrado, redondeado
datos['Valor m2'] = datos['Valor m2'].round(2)
datos.head(10)
```

```python
#cradno valor bruto m2 y redondeandolo en una sola variable
datos['Valor Bruto m2'] = ((datos['Valor Bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
#modificando tipos
casa = ['Casa', 'Casda en condominio', 'Casa de villa']
```

```python
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
datos
```
