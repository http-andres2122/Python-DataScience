```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
```

```python
datos.head(10)
```

```python
#creando lista pero filtrada
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = [
 'Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

```python
datos.head(10)
```

```python
#evaluando datos con el metodo .isin
depurando = datos['Tipo'].isin(residencial)
depurando
```

```python
datos_residencial = datos[depurando]
```

```python
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
#corrigiendo indices
datos_residencial.shape
```

```python
datos.shape
```

```python
datos_residencial.index = range(datos_residencial.shape[0])
```

```python
datos_residencial
```

#Exportando base de datos
