#Traducciendo variables

```python
#cambiando nombre de variables
```

```python
datos.columns
```

```python
dic_columnas = {
    'LIMIT_BAL' : 'Limite',
    'CHECKING_ACCOUNT' : 'Cuenta_corriente',
    'EDUCATION' : 'Escolaridad',
    'MARRIAGE' : 'Estado_civil',
    'AGE' : 'Edad',
    'BILL_AMT' : 'Valor_factura',
    'PAY_AMT' : 'Valor_pago',
    'DEFAULT' : 'Moroso'
}
```

```python
tarjetas = datos.rename(columns = dic_columnas)
```

```python
tarjetas.head()
```

##cambiando variables dentro de las columnas

```python
#llamando valores unicos en columna
tarjetas.Cuenta_corriente.unique()
```

```python
dic_cuenta = {
    'Yes' : 'Si' ,
    'No' : 'No'
}
```

```python
#generando valor sin atribuir
tarjetas.Cuenta_corriente.map(dic_cuenta)
```

```python
#atribuyendo valores a columna
tarjetas.Cuenta_corriente = tarjetas.Cuenta_corriente.map(dic_cuenta)
```

```python
tarjetas.head()
```

```python
#llamando valores unicos de escolaridad
tarjetas.Escolaridad.unique()
```

```python
dic_Escolaridad = {
    '2.University' : '2. Universidad',
    '3.Graduate School' : '3. Post-graduacion',
    '1.High School' : '1. Colegio'
}
```

```python
tarjetas.Escolaridad = tarjetas.Escolaridad.map(dic_Escolaridad)
```

```python
tarjetas.head()
```

```python
#llamando datos unicos estado civil
tarjetas.Estado_civil.unique()
```

```python
dic_estado_civil = {
    'Married' : 'Casado/a',
    'Single' : 'Soltero/a'
}
```

```python
tarjetas.Estado_civil = tarjetas.Estado_civil.map(dic_estado_civil)
```

```python
tarjetas.head(10)
```

```python
tarjetas.to_csv('/content/creditos_finales.csv', sep=';', index=False)
```

#Test de hipÃ³tesis

```python
#test de hipotesis con varibale morosidad
sns.lmplot(x='ui', y='Valor_factura', data=tarjetas, col='Moroso')
```

```python
#biblioteca scipy
from scipy.stats import ranksums
```

```python
tarjetas.head()
```

H<sup>null<sup>

la distribucion de los grupos moroso y no moroso es la misma

H<sup>alt<sup>

la distribucion de los grupos moroso y no moroso no es la misma

```python
#creando grupo moroso y no moroso
moroso = tarjetas.query('Moroso == 1').Valor_factura
moroso
```

```python
no_moroso = tarjetas.query('Moroso == 0').Valor_factura
no_moroso
```

```python
#funcion ranksums
#los valores abajo de 5% se acepta como hipotesis alternativa aceptada
ranksums(moroso, no_moroso)
```

```python
resultado = ranksums(moroso, no_moroso)
```

```python
print('el p_value de nuestro test es {}'.format(resultado.pvalue))
```
