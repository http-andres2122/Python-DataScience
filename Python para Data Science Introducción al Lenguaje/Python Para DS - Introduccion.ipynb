
#Bucles y Ciclos

```python
## bucle fuera de la funcion
edades = [18, 15, 12, 21, 45, 50]

def verificar_licencia(edad):
  if edad >= 18:
    print('Usted SI puede coducir')
  else:
    print('Usted NO puede conducir')


for edad in edades:
  verificar_licencia(edad)
```

```python
## bucle dentro de la funcion
edades = [18, 15, 12, 21, 45, 50]

def verificar_licencia_con_bucle(edades):
  for edad in edades:
    if edad >= 18:
      print('Usted SI puede coducir')
    else:
      print('Usted NO puede conducir')

verificar_licencia_con_bucle(edades)
```

```python
#ejempl alura
def clasificar_nps(notas):
  for nota in notas:
    if nota <= 6:
      print('El alumno es considerado Detractor')
    else:
      if nota <= 8:
        print('El alumno es considerado Neutro')
      else:
        print('El alumno es considerado Promotor')

clasificar_nps([9,8,10,6,7])
```

```python
verificaciones = []
edades = [12, 14, 20, 35]

def verificar_licencia_bool(verificaciones, edades):
  for edad in edades:
    if edad >= 18:
      verificaciones.append(True)
    else:
      verificaciones.append(False)
  for verificacion in verificaciones:
    if verificacion == True:
      print("Usted SI puede conducir")
    else:
      print("Usted NO puede conducir")

verificar_licencia_bool(verificaciones, edades)
```

```python
#ejemplo alura
a = 2

if a < 2:
 a = a + 1
else:
 a = a + 2
if a < 4:
 a = a - 4
else:
 a = a + 6

a

#explicacion
#¡Correcto! En la primera condición if se comprueba si el contenido de la variable a es menor que 2.
#El valor almacenado en a es 2. Por lo tanto, podemos afirmar que 2 no es menor que 2.
#Continuando con la sentencia condicional, el proceso entra en el else.
#Entonces, la variable a es sumada de 2 y acaba atribuyendo el valor 4. Luego en la secuencia, en el segundo if verifica si a es menor que 4.
#Nuevamente, podemos afirmar que 4 no es menor que 4. Por lo tanto, el flujo entra al segundo else y en este caso acaba sumando 6 al 4 que ya tenía a. Es así que, el valor final es 10.
```
