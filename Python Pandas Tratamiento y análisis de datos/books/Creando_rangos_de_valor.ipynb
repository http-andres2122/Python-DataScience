```python
import pandas as pd
```

```python
data = pd.read_csv('alquiler.csv', sep = ';')
```

```python
data.head(10)
```

```python
#agrupando objetos s
# 1 y 2
# 3 y 4
# 5 y 6
# 7 o mas cuartos
clases = [0, 2, 4, 6, 100]
```

```python
#creando cortes
cuartos = pd.cut(data.Cuartos, clases)
```

```python
cuartos
```

```python
#visualizar frecuencias
pd.value_counts(cuartos)
```

```python
#generando etiquetas para frecuencias
label = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 cuartos o mas']
```

```python
cuartos = pd.cut(data.Cuartos, clases, labels = label)
cuartos
```

```python
#generar frecuencia
pd.value_counts(cuartos)
```

```python
#incluyendo 0
cuartos = pd.cut(data.Cuartos, clases, labels = label, include_lowest = True)
cuartos
```

```python
#generar frecuencia
pd.value_counts(cuartos)
```
