```python
datos['Distrito'].drop_duplicates()
```

```python
#guardando en data frame
grupo_barrio = datos.groupby('Distrito')
```

```python
type(grupo_barrio)
```

```python
grupo_barrio.groups
```

```python
#iterador para recorer objeto grupo barrio
for barrio, data in grupo_barrio:
  print('{} -> {}'.format(barrio, data.Valor.mean()))
#crea un data frame, con cada iteracion del grupo
```

```python
#metodos mas simples de hacer lo de arriba
grupo_barrio['Valor'].mean()
```

```python
#redodeando decimales
grupo_barrio['Valor'].mean().round(2)
```

```python
#con mas de una columna
grupo_barrio[['Valor', 'Mantenimiento']].mean().round(2)
```

#generando estadisticas descriptivas

```python
grupo_barrio['Valor'].describe().round(2)
```

```python
#mostrando solo filas que se quieren
grupo_barrio['Valor'].aggregate(['min', 'max'])
```

```python
#renombrando columnas
grupo_barrio['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Minimo', 'max': 'Maximo'})
```

```python
#biblioteca que genera graficos
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_barrio['Valor'].mean().plot.bar(color = 'orange')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor medio del alquiler por distrito', {'fontsize':22})
```

```python
fig = grupo_barrio['Valor'].max().plot.bar(color = 'orange')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor maximo del alquiler por distrito', {'fontsize':22})
```
