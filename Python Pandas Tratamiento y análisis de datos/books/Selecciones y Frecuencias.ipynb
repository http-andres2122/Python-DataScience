#Reporte de analisis V

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv',sep = ';')
```

```python
datos.head(10)
```

```python
#Seleccione solamente los inmuebles clasificados como tipo 'Departamennto'
```

```python
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0]
n1
```

```python
n1.shape[0]
```

```python
#Seleccione los inmuebles clasificados como tipo 'Casa', 'Casa en condominio' y 'Casa de villa'
```

```python
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
```

```python
n2 = datos[seleccion].shape[0]
n2
```

```python
#Seleccione los inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites
```

```python
seleccion = (datos['Area'] >=60) & (datos['Area'] <=100)
n3 = datos[seleccion].shape[0]
n3
```

```python
n3
```

```python
#Seleccione los inmuebles que tengan por lo menos 4 cuarts y alquiler menor que $2.000,00
```

```python
seleccion = (datos['Cuartos'] >=4) & (datos['Valor'] <=2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print("N° de inmuebles clasificados como tipo 'Departamennto' -> {}".format(n1))
print("N° de inmuebles clasificados como tipo 'Casa', 'Casa en condominio' y 'Casa de villa' -> {}".format(n2))
print("N° de inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites' -> {}".format(n3))
print("N° de inmuebles que tengan por lo menos 4 cuarts y alquiler menor que $2.000,00 -> {}".format(n4))
```
