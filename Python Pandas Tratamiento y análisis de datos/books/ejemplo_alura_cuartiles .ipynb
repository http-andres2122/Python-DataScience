```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14, 6))
```

```python
datos = pd.read_csv('alquiler_muestra_alquiler_muestra.csv', sep = ';')
datos.head(10)
```

```python
#creando box-plot
datos.boxplot(['Valor m2'])
```

```python
#buscando datos mayores, se toma la medida donde se encuentre la mayoria de datos
datos[datos['Valor m2'] >= 2000]
```

```python
#variable valor
valor = datos['Valor m2']
```

```python
#ejecutando cuartiles
#ejemplo buscar
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1, Q3, IIQ.round(2), limite_inferior.round(2), limite_superior
```

```python
seleccion = ((valor >= limite_inferior) & (valor <= limite_superior))
```

```python
datos_new = datos[seleccion]
```

```python
#generando box plot con cuartiles de tabla nueva
datos_new.boxplot(['Valor m2'])
```

```python
#generando historico con tabla nueva
datos_new.hist(['Valor m2'])
```

```python
#generando historico con tabla antigua
datos.hist(['Valor m2'])
```
