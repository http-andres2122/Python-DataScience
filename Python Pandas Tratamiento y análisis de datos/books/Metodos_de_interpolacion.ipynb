```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
#llenar valores nulos con 0,metodo fillna, si no se coloca Inplace = True no ejecuta cambios en la tabla
s.fillna(0)
```

```python
s
```

```python
#metodo ffill, si encuenta un na lo sustituye por su valor anterior de arriba a abajo
s.fillna(method = 'ffill')
```

```python
s
```

```python
#metodo bfill, si encuentra un na lo sustituye por su valor anterior pero de abajo hacia arriba
s.fillna(method = 'bfill')
```

```python
s
```

```python
#calcular media aritmetrica, y sustituir los na
s.fillna(s.mean())
```

```python
s
```

```python
#limitando llenado de casillas na para evitar rectas a la hora de graficar
s.fillna(method = 'ffill', limit = 1)
```

```python
s
```

```python
#combinando metodos para llenar casillas NA,combinando de arriba a bajo
s1 = s.fillna(method = 'ffill', limit = 1)
s1
```

```python
#camnbiamos s por s1
s2 = s1.fillna(method = 'bfill', limit = 1)
s2
```
